Demetlerin listelerden farkı, içindeki elemanların değiştirilemiyor olmasıdır. Demetler, Read Only olduğu için listelere göre daha hızlı çalışabilir.

Eleman değiştirilemez, eklenemez, silinemez. Ancak aynı değişken tekrar tanımlanabilir.

Sadece iki method kullanılabilir; index ve count

```
demet = (1,2,3,4,5)
type(demet)
```

```
demet.index(3)
```

```
demet = (2,3,4,5,5,5,5,6)
print(demet)
demet.index(5)
```

```
demet.count(5)
```
