```
print("ax^2 + bx + c  olacak şekilde a,b ve c değerlerini giriniz:")
a=int(input("a="))
b=int(input("b="))
c=int(input("c="))

d=float((b**2)-(4*a*c))
if d<0:
  print("reel sayılarda çözümü yoktur.")
else:
  if d==0:
    k1= (-b - (d**0.5))/(2*a)
    print("\n\n{}x^2+{}x+{}\nÇ.K.={}".format(a,b,c,k1))
  else:
    k1= (-b - (d**0.5))/(2*a)
    k2= (-b + (d**0.5))/(2*a)

    print("\n\n{}x^2+{}x+{}\nÇ.K=({:.2f},{:.2f})".format(a,b,c,k1,k2))
```
