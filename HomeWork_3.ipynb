```
def disp_shape(n):
  for i in range(1,n+1):
    print((i-1)*" ",end="")
    print(*range(i,n+1),sep=" ")

  for i in range(n-1,0,-1):
    print((i-1)*" ",end="")
    print(*range(i,n+1),sep=" ")

disp_shape(5)
```

```
def crown(n):
  for s in range(1,n+1):                            #satır döngüsü
    
    print(*range(1,s+1),2*(n-s)*" ",sep="",end="")  # ilk üçgen ve sonraki boşluklar

    print(*range(s,0,-1),sep="",end="")             # orta üçgen 1e kadar sayılar
    print(*range(2,s+1),2*(n-s)*" ",sep="",end="")  # orta üçgen 1den sonrası ve boşluklar

    print(*range(s,0,-1),sep="")                    # son üçgen

crown(5)
```

```
def is_armstrong(a):
  s=a
  count = 0
  while s: 
    count += (s % 10)**3
    s //= 10
  return count == a

def is_armstrong2(n):   # kısaltılmış bir versiyonu
  return sum([int(digit)** 3 for digit in str(n)]) == n

def is_armstrong3(n):   # kısaltılmış bir versiyonu daha
  return sum(map(lambda a : int(a)**3 str(n))) == n

def print_armstrong(a):
  for i in range(1,a+1):
    if is_armstrong(i):
      print(i,end=" ")

print_armstrong(1000000)

while True:
  a = int(input("bir sayı giriniz:"))
  if a==0:
    break
  print(a,"sayısı","Armstrong sayıdır" if is_armstrong(a) else "Armstrong sayısı değildir.")
```

```
def digitate(vals):
  result = []
  for i in range(len(vals)):
    basamaklar = []
    while vals[i]: 
      b = vals[i] % 10
      basamaklar.append(b)
      vals[i] //= 10
    demet= tuple(basamaklar)
    result.append(demet)
  return result

val = [23,4,765,34589,42]
result = digitate(val)
print(result)
```

```
def newton_pi(k):
  def fact(f):
    result=1
    for i in range(1,f+1):
      result *= i
    return result
  count=0
  for i in range(k+1):
    pay = (2**(i+1))*(fact(i)**2)
    payda = fact((2*i)+1)
    count += pay/payda
  return count

print(newton_pi(9))

def somayaji_pi(n):
  count = 3
  for i in range(n+1):
    pay = (-1)**i
    payda = (((2*i)+3)**3)-((2*i)+3)
    count += 4*pay/payda
  return count

print(somayaji_pi(5))

def bailey_borwein_plouffe_pi(k):
  count = 0
  for i in range(k+1):
    part1 = (1/16)**i
    part2 = ((4/((8*i)+1))-(2/((8*i)+4))-(1/((8*i)+5))-(1/((8*i)+6)))
    count += part1*part2
  return count

bailey_borwein_plouffe_pi(5)
```

```
def sort_tuple_list(a):
  a.sort(key=lambda tup: tup[2])
  a.sort(key=lambda tup: tup[1])
  a.sort(key=lambda tup: tup[0])
  return a

a=[('Tom',19,80),('John',20,90),('Jony',18,93),('Jony',17,91),('John',20,85)]
sort_tuple_list(a)
print(a)
```
